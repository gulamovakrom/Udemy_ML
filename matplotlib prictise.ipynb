```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
plt.style.use('seaborn')
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)
scatter = ax0.scatter(x=over_50["age"],
            y=over_50["chol"],
            c=over_50["target"],
             cmap="winter")
ax0.set(title="heart & Chol",
        ylabel="Chol")
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(over_50["chol"].mean(),linestyle='--')
ax0.set_ylim([150,400])


scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")
ax1.set(title="heart & Chol",
        ylabel="Thalach")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(over_50["thalach"].mean(),linestyle='--')
ax1.set_ylim([0,250]),ax1.set_xlim([50,75])
fig.suptitle("Heart Disease Info", fontsize=16,fontweight="bold");
```
