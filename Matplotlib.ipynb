## Making figures with NumPy arrays

We want to:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots


```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
x[:10]
```

```python
fig,ax = plt.subplots()
ax.plot(x,x**2);
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x))
ax.set(title="Graph of sin(x)",
       ylabel="y-values",
       xlabel="x-values");
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter":10,
                     "Peanut buteer":8,
                     "Cashew butter":12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Dan's Butter Stote",
       ylabel="Price ($)");
```

```python
# Histograms
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# multiple options
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,8))
ax1.plot(x,x/2)
ax2.scatter(np.random.randn(10),np.random.randn(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax3.set(title="Dan's Butter Store", ylabel="Price ($)")
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      figsize=(10,8))
ax[0].scatter(np.random.randn(10),np.random.randn(10))
ax[1].plot(x,x/2);
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(nrows=2,
                                          ncols=2,figsize=(10,5))
ax[0,0].scatter(np.random.randn(10),np.random.randn(10)),ax[0,1].bar([1,2,3],[10,20,30]),ax[1,0].plot([1,2,3],[1,1,1]),ax[1,1].hist(np.random.randn(10000));
```
